```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

import scienceplots  
plt.style.use(['science','notebook', 'grid'])

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : '14'}
rc('font', **font)
```

```python
height = np.array([0.09, 0.08, 0.07, 0.06, 0.05, 0.04, 0.03, 0.02, 0.01])
T_sq = np.array([0.00138 , 0.001268, 0.001134, 0.001008, 0.000883, 0.000741,
       0.00062 , 0.00049 , 0.000384])
# Delta T^2
yerr = np.array([0.00002, 0.000005, 0.000005, 0.000005, 0.000007, 0.000005, 0.000005, 0.000006, 0.000008])

fake_data = np.linspace(0.00038, 0.00138, 9)
fake_data = fake_data[::-1]
print(fake_data)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.errorbar(height, T_sq, yerr = yerr, fmt = 'none', capsize=3)
ax.plot(height, fake_data, '--', linewidth=1)
ax.set_xlabel('Height [cm]')
ax.set_ylabel('$T^2$ $[s^2]$')
ax.text(0.35, 0.1, '$(0.0127x \pm 0.0001) + (0.000244\pm 0.000006)$',
        transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='blue'))
# plt.savefig('fig1.png')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.errorbar(height, T_sq, yerr = yerr, fmt = 'none', capsize=3)
ax.plot(height, fake_data, '--', linewidth=1)
ax.set_xlabel('Height [m]')
ax.set_ylabel('$T^2$ $[s^2]$')
ax.text(0.55, 0.1, '$y = 0.0127(1)x + 0.000244(6)$',
        transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='blue'))
plt.savefig('fig1.png')
plt.show()
```
