```python
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# x-data
logT = np.array([-0.06, 0.1, 0.18, 0.23, 0.269, 0.302, 0.329, 0.352, 0.371, 0.389])
# y-data
logP = np.array([-2.093, -1.656, -1.336, -1.117, -0.944, -0.81, -0.696, -0.601, -0.519, -0.445])
yerr = np.array([0.08, 0.03, 0.01, 0.008, 0.005, 0.004, 0.003, 0.002, 0.002, 0.002]
)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(logT, logP)

# Create a line of best fit
best_fit_line = slope * logT + intercept

# Create the plot and get the axis
fig, ax = plt.subplots()

# Plot the data points
ax.scatter(logT, logP, color='blue', label='Data points')

# Plot the best fit line
ax.plot(logT, best_fit_line, color='red')

# Plot the error bars
ax.errorbar(logT, logP, yerr=yerr, ecolor='k', fmt='none', capsize=3)

# Add labels and title
ax.set_xlabel('log(T)')
ax.set_ylabel('log(P)')

# Add text with the equation of the line
ax.text(0.10, 0.85, '$y = 3.8(1)x-1.94(3)$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='blue'))

plt.xlim(-0.1)

plt.savefig('fig1.png')
# Show the plot
plt.show()
```
