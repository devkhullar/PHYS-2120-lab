```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

import scienceplots  
plt.style.use(['science','notebook', 'grid'])

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : '14'}
rc('font', **font)
```

I am creating a plot of $\frac{1}{m\lambda}$ vs $\frac{1}{m}\left(\frac{1}{n^2_{f}}-\frac{1}{n^2_{i}} \right)$

```python
# Data
x = np.array([0.2100, 0.1875, 0.1389, 0.1050, 0.0938, 0.0694])
y = np.array([2279962.1212, 2085480.9134, 1533418.7579, 1155743.4145, 1032151.1974, 770125.2816])
yerr = np.array([3498.353532, 2906.379988, 1512.024935, 800.0819647, 608.8447509, 265.3273786])

# Perform a linear fit
coeffs = np.polyfit(x, y, 1)  # Linear fit (degree 1)
m, c = coeffs  # Slope and intercept

# Generate best-fit line
x_fit = np.linspace(min(x), max(x), 500)
y_fit = m * x_fit + c

# Plotting
plt.errorbar(x, y, yerr=yerr, fmt='none', capsize=5)
plt.plot(x_fit, y_fit, label=r'$y = (1.1 \pm 0.1) \times 10^7 x + (1 \pm 2) \times 10^4$', color='red')
plt.ylabel(r'$\frac{1}{m \lambda} \, [10^6 \, \mathrm{m^{-1}}]$', fontsize=20)  # Add *10^6 to y-axis label
plt.xlabel(r'$\frac{1}{n_f^2} - \frac{1}{n_i^2}$', fontsize=20)

# Remove the scientific notation offset at the top of the plot
plt.gca().get_yaxis().get_offset_text().set_visible(False)

plt.legend()
plt.grid(True)
plt.savefig('atomic-spectra.jpg')
plt.show()
```
