```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

import scienceplots  
plt.style.use(['science','notebook', 'grid'])

from matplotlib import rc
rc('text', usetex=True)
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : '14'}
rc('font', **font)
```

```python
# x-data
h = np.array([0, -0.0108, 0.0252, 0.0379, 0]) * 100  # height
# y-data
p = np.array([90.5, 93.1, 93.07, 90.49, 90.5]) # pressure
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
colors = ['red','blue', 'green', 'purple', 'red']
ax.plot(h, p, '-')
ax.scatter(h, p, color=colors)
ax.set_ylabel("Absolute Pressure [kPa]")
ax.set_xlabel("Height [m]")
ax.set_xlim(-1.7, 4.3)
ax.text(0.22, 0.075, '$A$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='red'))
ax.text(0.05, 0.91, '$B$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='blue'))
ax.text(0.75, 0.91, '$C$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='green'))
ax.text(0.94, 0.075, '$D$', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='purple'))
plt.savefig('fig1')
plt.show()
```
